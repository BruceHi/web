```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dates = pd.date_range('20200101', periods=12)
df = pd.DataFrame(np.random.randn(12, 4), index=dates, columns=list('ABCD'))
df
```

```python
# 要安装 xlwt 软件包
df.to_excel(excel_writer='file.xlsx', sheet_name='ok', index=False,
            columns=['A', 'C'])
```

```python
plt.plot(df.index, df['A'])
plt.show()
```

```python
plt.plot(df.index, df['A'], color='#FFAA00', linestyle='--',
         linewidth=3, marker='D')
plt.show()
```

```python
# 散点图
plt.scatter(df.index, df['A'])
plt.show()
```

```python
sns.set_style('darkgrid')
plt.scatter(df.index, df['A'])
plt.show()
```
