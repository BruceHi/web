```python
import re
import pandas as pd
```

```python
with open('softmax.txt', 'r') as f:
    s = f.read()
s
```

```python
p = r'(?<= accuracy: )\d.\d+'
pattern = re.compile(p)
acc = pattern.findall(s)
acc = [float(x) for x in acc]
acc
```

```python
p = r'(?<= val_accuracy: )\d.\d+'
pattern = re.compile(p)
val_acc = pattern.findall(s)
val_acc = [float(x) for x in val_acc]
val_acc
```

```python
with open('center_loss.txt', 'r') as f:
    s2 = f.read()
s2
```

```python
p = r'(?<= dense_7_accuracy: )\d.\d+'
pattern = re.compile(p)
acc_cen = pattern.findall(s2)
acc_cen = [float(x) for x in acc_cen]
acc_cen
```

```python
p = r'(?<= val_dense_7_accuracy: )\d.\d+'
pattern = re.compile(p)
val_cen = pattern.findall(s2)
print(len(val_cen))
val_cen = [float(x) for x in val_cen]
val_cen
```
